# Scraping Data Ulasan Aplikasi Duolingo dari Google Play Store

Notebook ini berisi proses pengambilan data ulasan aplikasi **Duolingo** (`com.duolingo`) dari Google Play Store menggunakan pustaka `google-play-scraper`. Data yang diambil kemudian disimpan dalam format `.csv` untuk digunakan pada tahap analisis sentimen selanjutnya.

## 1. Instalasi Library

```python
!pip install google-play-scraper -q

import pandas as pd
from google_play_scraper import reviews, Sort
```

## 2. Proses Scraping Ulasan

Mengambil ulasan aplikasi Duolingo (ID aplikasi: `com.duolingo`) berbahasa Indonesia dari Google Play Store, diurutkan berdasarkan relevansi (`Sort.MOST_RELEVANT`), sebanyak maksimal 15.000 ulasan.

```python
scrapreview, _ = reviews(
    'com.duolingo',          # ID aplikasi Duolingo di Google Play Store
    lang='id',                # Bahasa ulasan: Indonesia
    country='id',             # Negara: Indonesia
    sort=Sort.MOST_RELEVANT,  # Urutan ulasan berdasarkan relevansi
    count=15000                # Jumlah maksimum ulasan yang diambil
)

print(f"Jumlah ulasan berhasil diambil: {len(scrapreview)}")
```

## 3. Konversi ke DataFrame dan Pembersihan Duplikat Awal

```python
app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df = app_reviews_df.drop_duplicates(subset='reviewId')

print(f"Shape data setelah drop duplikat reviewId: {app_reviews_df.shape}")
app_reviews_df.head()
```

## 4. Menyimpan Dataset ke CSV

```python
app_reviews_df.to_csv('ulasan_aplikasi.csv', index=False)
print("Dataset berhasil disimpan sebagai 'ulasan_aplikasi.csv'")
```

Ringkasan struktur dan jumlah data hasil scraping.
